Food Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("Cleaned_Indian_Food_Dataset.csv")

print(df.shape)
df.head()
```

```python
df.TranslatedInstructions[0]
```

```python
import re
text = df.TranslatedInstructions[0]
```

```python
def preprocess(text):
    text = re.sub(r"[^\w\s']", "", text)
    text = re.sub(r"[\n]+", "", text)
    text = text.strip().lower()
    return text
```

```python
preprocess(text)
```

Met de function map kan je de hele kolom tranformeren

```python
df.TranslatedInstructions = df.TranslatedInstructions.map(preprocess)
```

```python
df.TranslatedInstructions[0]
```

CBOW AND SKIPGRAM zijn unsupervised learning

```python
df.to_csv("food_dataset.txt", columns=["TranslatedInstructions"], header=None, index=False)
```

```python
with open("food_dataset.txt", "r") as f:
    exported = f.read()

print(exported)
```

```python
import fasttext

model = fasttext.train_unsupervised("food_dataset.txt")
```
